```python
## Question 1
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Colab/123.csv")
df_new = df.dropna(subset=['Labour force participation rate, female (% ages 15 and older) (1994)'])
df.describe
```

```python
theta_0,tau = 0,1; alpha,beta = 1,0.2
import pymc as pm
normal_gamma_force_model = pm.Model()
with normal_gamma_force_model:
    theta = pm.Normal("theta", mu=0, sigma=1)
    phi = pm.Gamma("phi", alpha=1, beta=1)
    x_obs = pm.Normal("likelihood", mu=theta, sigma=1/phi**0.5, observed=df_new["Labour force participation rate, female (% ages 15 and older) (1994)"])
```

```python
with normal_gamma_force_model:
    idata_HMC = pm.sample()
```

```python
import numpy as np
with normal_gamma_force_model:
    MH = pm.Metropolis([theta, phi], S=np.array([0.1]), tune=False, tune_interval=0)
    idata_MH = pm.sample(step=MH)
```

```python
import arviz as az
import matplotlib.pyplot as plt
fig,ax = plt.subplots(2,2, figsize=(14,3))
az.plot_trace(idata_HMC, axes=ax[:2,:2])
print("HMC Split-Rhats", {k: np.round(v.values,4) for k,v in az.rhat(idata_HMC).items()}, end=' | ')
```

```python
import arviz as az; print("number of MCMC samples:", idata_HMC.posterior['phi'].values.flatten().shape[-1])
az.summary(idata_HMC, round_to=2)
```

```python
import arviz as az; import matplotlib.pyplot as plt; fig,ax = plt.subplots(1,2, figsize=(10,6))
az.plot_forest(idata_HMC, var_names=["theta","phi"], combined=False, hdi_prob=0.95, ax=ax[0])
```
