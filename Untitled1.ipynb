```python
# Q2, part 1
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
# Parameters
n = 3
theta_0, tau = 4, 1/2
parameter_domain = np.linspace(-10, 10, 1001)
# individual likelihood
x = stats.norm(loc=0,scale=1).rvs(n)
likelihoods = stats.norm.pdf((x[:,np.newaxis]), loc=parameter_domain, scale=1) # sigma=1 known
prior = stats.norm(loc=theta_0, scale=1/tau**0.5).pdf(parameter_domain) 
c,c2 = 500,10 
joint = c*np.prod(likelihoods)*prior # elementwise multiplication 
posterior = stats.norm(loc=(tau*theta_0+x.sum())/(tau+n), scale=1/(tau+n)**0.5).pdf(parameter_domain)
plt.figure(figsize=(16,5))
x=stats.norm(loc=0,scale=1).rvs(n)
for xi in x:
    likelihoods = stats.norm.pdf(xi, loc=parameter_domain, scale=1) # sigma=1 known
    plt.plot(parameter_domain, c2*likelihoods, label = "likelihood")
plt.plot(parameter_domain, prior, label="prior")
plt.plot(parameter_domain, posterior, label="posterior")
plt.plot(parameter_domain, joint, linestyle="--", label="joint") 
plt.legend()
display(c2*likelihoods)
```

```python
# Q2 part 2:
# Q2, part 1
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
# Parameters
n = 3
theta_0, tau = 4, 1/2
parameter_domain = np.linspace(-10, 10, 1001)
x = stats.norm(loc=0,scale=1).rvs(n)
likelihoods = stats.norm.logpdf((x[:,np.newaxis]), loc=parameter_domain, scale=1) # sigma=1 known
prior = stats.norm(loc=theta_0, scale=1/tau**0.5).logpdf(parameter_domain) 
c,c2 = 500,10 
joint = c*np.prod(likelihoods)*prior # elementwise multiplication 
posterior = stats.norm(loc=(tau*theta_0+x.sum())/(tau+n), scale=1/(tau+n)**0.5).logpdf(parameter_domain)
plt.figure(figsize=(16,5))
x=stats.norm(loc=0,scale=1).rvs(n)
for xi in x:
    likelihoods = stats.norm.logpdf(xi, loc=parameter_domain, scale=1) # sigma=1 known
    plt.plot(parameter_domain, c2*likelihoods)
plt.plot(parameter_domain, prior, label="prior")
plt.plot(parameter_domain, posterior, label="posterior")
plt.plot(parameter_domain, joint, linestyle="--", label="joint (times scalar "+str(c)+")") 
plt.legend()
display(c2*likelihoods)
```
